## Lojistik Regresyon Modelinin Kurulması 

Basit doğrusal regresyon modeli;

$$Y= \beta_0+\beta_1X+ε$$

Çoklu doğrusal regresyon modeli ise;

$$Y= \beta_0+\beta_1X_1+\beta_2X_2+...+\beta_kX_k + ε$$

ile ifade edilir. Burada;

$Y$: Bağımlı değişkeni

$X_1, X_2,..., X_k $: Bağımsız değişkenleri

$\beta_0$: Bağımsız değişkenler sıfır değerini aldığında bağımlı değişkenin değerini, yani sabiti

$\beta_1, \beta_2,..., \beta_k$ : Bağımsız değişkenlerin regresyon katsayılarını

$\varepsilon$: Hata terimini

$k$: Bağımsız değişken sayısını göstermektedir. 

Basit ve çoklu doğrusal regresyon modelinde, bağımlı değişkenin, verilen bağımsız değişkenlerin değerlerine göre beklenen değeri (ortalama değeri) basit doğrusal regresyon modeli için; 

$$E(Y|X)=\beta_0+\beta_1X_1$$

çoklu doğrusal regresyon modeli için ise;



$$E(Y|X_1,X_2,...,X_k)=\beta_0+\beta_1X_1+\beta_2X_2+...+\beta_kX_k$$

Lojistik regresyon modeli, genel doğrusal modellerin binom dağılımlı bağımlı değişkenler için elde edilmiş olan özel bir biçimidir. 
Lojistik fonksiyonun 0 ile 1 arasında bir değişim aralığına sahip olması lojistik fonksiyonun tercih edilmesindeki önemli nedenlerden biridir.

İncelenen bir olayın olasılığının kendi dışında kalan diğer olayların olasılığına oranına “odds değeri” denilmekte ve şu şekilde hesaplanmaktadır:



$$\text{Odds Değeri} = \frac{p}{1-p}=e^{ \alpha +\beta_x}$$

$$odds=\frac{p}{1-p}=\frac{\text{probability of presence of characteristic}}{\text{probability of absence of characteristic}}$$

Burada P üzerinde durulan olayın olasılığını, 1-P ise üzerinde durulmayan olayın olasılığını göstermektedir. Odds değeri 0
ile +∞ arasında değerler almaktadır.
İncelenen iki farklı olayın odds değerlerinin birbirine oranına “odds oranı (odds ratio, OR)” denilmektedir. Odds oranı, incelenen iki olayın gözlenme olasılıklarından birinin diğerine oranla kaç kat daha fazla veya kaç kat daha az olarak ortaya çıkabileceğini göstermekte olup şu eşitlik ile hesaplanmaktadır:



$$OR=\frac{e^\alpha+\beta}{e^\alpha}=e^\beta$$

İncelenen bir olasılığın (P), odds değerinin doğal logaritması lojit fonksiyon olarak adlandırılır.


$$logit(p)=ln\frac{p}{1-p}$$

İncelenen olasılığın odds değeri 0 ile +∞ arasında değer alırken aynı olasılığın lojit değeri -∞ ile +∞ arasında değerler alabilmektedir.


$$=log \frac{p(y=1)}{1-(p=1)} = \beta_0+\beta_1X_1+\beta_2X_2+...+\beta_mX_m $$

Olasılıkların lojit fonksiyonunun kullanılmasının amacı, doğrusal bir model elde
edilerek, parametre tahminlerinin yapılmasıdır. İncelenen bir olasılığın (P) lojit değeri
doğrusal modele eşitlendiğinde; 

$$logit(p) = b_0 + b_1X_1 + b_2X_2 + b_3X_3 +...+ b_kX_k $$

elde edilir. Elde edilen bu eşitlik, basit ve çoklu doğrusal regresyon
modellerindeki bağımlı değişkenin beklenen değerini veren ve parametre tahminlerinde
kullanılan eşitliğe benzer bir eşitliktir.

Son iki eşitlikten;
$$ p(x)= \frac{e^{\beta_0+\beta_1x_1+\beta_2x_2+...+\beta_nx_p}}{1+ e^{\beta_0+\beta_1x_1+\beta_2x_2+...+\beta_kx_k}}$$

elde edilir. Bu eşitliğe “lojistik regresyon modeli” denir. Karekök hataların toplamını en aza indirgeyen parametreleri seçmek yerine (sıradan regresyon gibi, en küçük kareler yöntemi), lojistik regresyonda tahmin, örnek değerlerin gözlem olasılığını en yükseğe çıkaran parametreleri seçer(maksimum olabilirlik yöntemi).

P : İncelenen olayın gözlenme olasılığını

$\beta_0$ : Bağımsız değişkenler sıfır değerini aldığında bağımlı değişkenin değerini, yani sabiti,

$\beta_1,\beta_2,…,\beta_n$ : Bağımsız değişkenlerin regresyon katsayılarını,

$X_1,X_2,…,X_n$ : Bağımsız değişkenleri

p : Bağımsız değişken sayısını

e = 2,718 sayısını göstermektedir. 

(https://veribilimcisi.com/2017/07/18/lojistik-regresyon/)

### (*) İşaretinin Anlamı:


Bir regresyon tablosundaki bir yıldız işareti "p <.1" anlamına gelmektedir. İki yıldız işareti "p <.05"; ve üç yıldız işareti 
"p <.01" anlamına gelmektedir. Regresyon tablosundaki yıldız işaretleri, bir regresyon katsayısının istatistiksel önem düzeyini göstermektedir. Örneğin, p <.05. bir sonuç elde etme olasılığı 0.05 veya daha düşüktür. Başka bir deyişle 100 örnekten 95 kez aynı pozitif, sıfır olmayan yani 1 sonucuna sahip etkiyi görmeyi bekleriz. Yani 0.05 in güveni %95, 0.01 in güveni %99'dur. Bu yüzden daha iyi bir sonuç almak için yeni modelimizi yıldızlı değişkenleri alarak kurarız. Sadece üç yıldızlıları aldığımız model gerçeğe en yakın tahminde bulunmalıdır.
(http://svmiller.com/blog/2014/08/reading-a-regression-table-a-guide-for-students/)

## Logistic Regression Modelinin Uygunluğunun ve Doğruluğunun Değerlendirilmesi:

Lojistik Regresyon için kullanılan değerlendirme kriterleri şu şekildedir:

### a) Akaike Information Criteria (AIC):

AIC, modelin verilere uyma yeteneğini de hesaba katan uygunluk iyiliğinin bir ölçüsüdür. Model uyumu için önemli bir göstergedir. Daha küçük daha AIC'ye sahip model daha iyi sonuç verir.

Bir modelin AIC metriğine bakmak modelleri karşılaştırmada oldukça kullanışlıdır. Dolayısıyla, oluşturulan modelin daha iyi sonuç vermesini yani daha düşük AIC değerine sahip olması istenirse, regresyon tablosundaki yıldızlı değişkenler seçilerek yeni modeller ( fraud için model ve model2'nin önemli değişkenleri alarak kurulması) kurulabilir. En düşük AIC'ye sahip model nispeten daha iyi olacaktır. 
(https://www.hackerearth.com/practice/machine-learning/machine-learning-algorithms/logistic-regression-analysis-r/tutorial/)

AIC, doğruluğu karmaşıklıkla dengeleyen bir metriktir: Daha yüksek doğruluk, daha iyisonuçlar verir, eklenen daha fazla değişken daha kötü sonuçların elde edilmesine yol açar. Düşük AIC'li model daha iyidir. (http://docs.statwing.com/a-user-friendly-guide-to-logistic-regression/)

### b) Confusion Matrix:

Confusion matrix, sınıflandırma(classification) modellerini değerlendirmek için sıklıkla kullanılan en önemli metriktir. Gördüğünüz gibi, konfüzyon matrisi gerçek ve tahmin edilen değerleri bir tablo biçiminde ölçerek "karışıklığı" önler. 
Bir confusion matris, data içindeki gerçek çıktılarla karşılaştırıldığında sınıflandırma modeliyle yapılan doğru ve yanlış tahminlerin sayısını gösterir.  
NxN Matrisi, burada N hedef değerlerin(sınıfların) sayısıdır.





Accuracy: Modelin tahmin edilen kesinliğini belirler.
        
        = (True Positives + True Negatives)/(True Positives + True Negatives + False Positives + False Negatives)

True Positive Rate(TPR): Tüm pozitif değerlerin arasından doğru kaç kere tahmin edildiğini gösterir.

        = TP/(TP + FN)  , TPR = 1 - FNR, TPR aynı zamanda Sensitivity veya Recall olarak adlandırılır.
        
False Positive Rate(FPR): Tüm negatif değerlerden kaç tane negatif değerin yanlış tahmin edildiğini gösterir.

        = FP/(FP + TN)  , FPR = 1 - TNR, 
        
True Negative Rate(TNR): Negatif değerlerin hepsinden kaç tanesinin doğru olarak tahmin edildiğini gösterir.

        = TN/(TN + FP)  , Specificity olarak da isimlendirilir.

False Negative Rate(FNR): Pozitif değerlerin hepsinden kaç tanesinin yanlış tahmin edildiğini gösterir.

        = FN/(FN + TP)

Precision: Tahmini pozitif değerlerin hepsinden kaç değerin gerçekten pozitif olduğunu gösterir.

        = TP/(TP + FP)
        
F Score: F skoru precision ve recall'un harmonik ortalamasıdır. 0 ile 1 arasında değer alır. Değeri daha yüksek F skoru, modelin daha iyi olduğunu gösterir.

        = 2[(precision*recall)/(precision+recall)]

(https://www.hackerearth.com/practice/machine-learning/machine-learning-algorithms/logistic-regression-analysis-r/tutorial/)


### c) R-Squared (R2):

Modelin tahmin doğruluğunun nicelleştirilmesi için kullanılar sayısal bir metriktir. 0 ve 1 arasında değişir. 0, modelin tahmini bir değeri olmadığı; 1 ise modelin her şeyi mükemmel bir şekilde öngördüğü anlamına gelir.




Örneğin, solda temsil edilen veriler sağdaki verilerden daha az doğru bir modele götürür. Dağılım çizgisi boyunca bir çizgi çizmeye çalıştığımızı varsayarsak; sağ grafikteki kırmızılar maviden neredeyse tamamen ayrı kısımlardayken, sol grafikte bunu yapmak zordur.

"İyi" bir r-squared'in sabit tanımı yoktur. Bazı durumlarda, herhangi bir etkiyi görmek ilginç olabilir, diğerlerinde modeliniz son derece doğru olmadığı sürece işe yaramaz olabilir. Herhangi bir değişken eklendiğinde, r-squared hareketi yükselir, bu nedenle mümkün olan en büyük kareyi elde etmek hedef değildir. Bunun yerine, modelin doğruluğunu (r-squared) metriği ile (genellikle, içindeki değişkenlerin sayısı) dengelemek istersiniz.
(http://docs.statwing.com/a-user-friendly-guide-to-logistic-regression/)

## Model Uyum Kriterleri: 

Lojistik regresyonun model uyumunun değerlendirilmesinde Olabilirlik Oran Ki-Kare(G test) ve Ki-Kare testleri baz alınmaktadır.  G test, Fisher’s Exact test ve Khi-kare testi, 2x2 çapraz tablolar için önerilen testlerdendir. G testi, çapraz tabloların hücrelerindeki gözlenen frekansların, beklenenlere göre uyum iyiliğini ya da  diğer  bir  ifadeyle  ele  alınan  iki  faktörün birbirinden bağımsız olup olmadığını test etmektedir.

G(Olabilirlik  Oran  Ki-Kare)  testi, özel bir hipotezle gözlenen frekanslar ile beklenen frekansları karşılaştırmak amacıyla Ki-kare istatistiğine alternatif bir testtir. Ayrıca, teorik  olarak  daha sağlam  temellere  dayanması, Ki-kare dağılımına uygun olması, hücredeki beklenen frekansların 5'ten küçük olduğu durumlarda bile yapılabilmesi ve kolayca hesaplanabilmesi gibi avantajları olan G-testi yukarıda bahsedilen diğer iki test tekniği yerine alternatif olabilmektedir.

Lojistik  regresyonda G-testi(-2logL) ile Pearson  Ki-kare(Chi-Square) testi kullanılmaktadır. Lojistik regresyonda ilk aşama; modelin dataya olan uyumunun değerlendirilmesi olmalıdır. Buna ilişkin hipotez; H0: Model dataya uyumludur ya da  H1: Model dataya uyumlu değildir şeklinde kurulabilir. (https://sedatsen.files.wordpress.com/2016/11/10-sunum.pdf)


#### 1) Pearson Ki-kare Testi: 

Ki-Kare testi her bir kategori çiftine düşen frekans sayısı ile bu durumlara şansla düşebilecek frekans sayılarının karşılaştırılmasına dayanır. Gözlenen frekans ile beklenen frekans karşılaştırması diyebiliriz. Büyük örneklemlerde çok iyi yaklaşıma sahiptir.
Özellikle ki-kare testi yapabilmek için çaprazlık tablosundaki her hücrede 5’ten küçük frekans değerleri bulunmamalıdır. Bu da ki-karenin küçük örneklemlerde tercih edilmemesine neden olmuştur.




$$χ^2=\sum {\frac{(Gözlenen_{ij}-Model_{ij})^2}{Model_{ij}} }$$


#### 2) Fisher’s Exact Test (Fisher Kesin Olasılık Testi) :

Küçük örneklemlerde Pearson Ki-kare testinin tercih edilememesinden dolayı alternatif olarak ki-kareye göre daha doğru sonuçlar sunan Fisher Kesin Olasılık Testi geliştirilmiştir.Bu istatistik özellikle küçük örneklemlerden elde edilen 2x2 tabloları için kullanılsa da büyük örneklemlerden elde edilen diğer büyük boyuttaki tablolar için de kullanılabilir (analizler daha fazla zaman alabilir).

#### 3) Score  İstatistiği: 

Temeli olasılık fonksiyonuna dayanmayan bu istatistik, asimptotik olarak p serbestlik dereceli Ki-kare dağılımı göstermektedir. Bu istatistik, bağımlı değişken ile bağımsız değişken arasında ilişki olup olmadığı konusunda fikir vermektedir.

#### 4) Wald İstatistiği: 

Modelde yer alan bağımsız değişkenlere ait katsayıların önemli olup olmadığını değerlendirmede kullanılan bir istatistiktir. Student-t testi ile hesaplanan bu istatistik, örneklem genişliği arttıkça standart normal dağılım gösterir. Örneğin 20 serbestlik derecesi için % 95 çift-taraflı kritik değeri 2.09 iken, 100 serbestlik derecesi için bu değer 1.98 olmaktadır.
Bu istatistik, 

$$z = \frac{\beta_j}{\sqrt{var(\beta_j)}}  $$

şeklinde hesaplanır. Güven aralığı ise;

$$\beta_j = \pm \text{ } z_{1-\alpha/2}{\sqrt{var(\beta_j)}} $$

#### 5) Yates Düzeltmesi (Yates’s Correction):

2x2 çaprazlık tablolarında Pearson ki-kare değeri küçük p değerleri sunarak anlamlı değerler üretmeye eğilimlidir. Bu da I.Tür hata yapılma şansını artırır. Bu sorunu çözmek için Yates bir düzeltme önermiştir. Aşağıdaki formülün Pearson ki-kareden tek farkı pay kısmındaki gözlenen ile model farklarından 0.5 çıkarılmasıdır. Buradaki bulunan değer de Pearson ki-kare değeri
gibi yorumlanabilir (p<0.05).




$$\chi^2=\sum {\frac{(|Gözlenen_{ij}-Model_{ij}|-0.5)^2}{Model_{ij}} }$$


## Ki-Kare Testinin Varsayımları:

Verilerin bağımsızlığı: Verilerin tahmini sonuçları çaprazlık tablosunun sadece bir hücresine ait olabilir. Yukarıdaki veri analizi istatistikleri 2 kategorik değişken içeren durumlar için kullanılmaktadır. Bu 2 değişkenin kategori sayısına göre tablolarımız 2x2, 2x3, 3x3 vb… şeklinde adlandırılmaktadır. İki kategorik değişkenin olduğu durumlarda kikare ve diğer istatistikler hesaplanabilir. Eğer veride ikiden fazla kategorik değişken varsa loglinear modeller kullanılabilir. Loglinear modeller iki kategorik değişkenin olduğu veriler için de kullanılabilir.

### Log-lineer Model:

Burada da ana etki değişkenleri ve bu ana etkilerin etkileşimleri modele bağımsız değişken olarak girmektedir. Burada frekanslar üzerinden analizler yapıldığı için logaritma alınarak analizler gerçekleştirilmektedir.

### Log-lineer Model Varsayımları:

Ki-karenin bir uzantısı olan log-doğrusal modeller de ki-karede olduğu gibi frekans sayıları üzerinde bazı şartları gerektirir.
Çaprazlık tablosunun her hücresi 5’ten büyük frekansa sahip olmalı. Log-doğrusal modellerin güvenilir sonuçlar vermesi açısından tablodaki her hücrede 1’den küçük hiç değer olmaması ve 5’ten küçük frekansa sahip hücrelerin verinin %20’sini geçmemesi gerekir.

# Analysis of Variance (ANOVA)

Varyans analizi (ANOVA), bir veri setinde bulunan toplam değişkenliği iki kısma ayıran istatistikte kullanılan bir analiz aracıdır: sistematik faktörler ve rasgele faktörler. Sistematik faktörler, verilen veri kümesi üzerinde istatistiksel bir etkiye sahiptir, ancak rastgele faktörler yoktur. Analistler, bağımsız değişkenlerin bir regresyon çalışmasında bağımlı değişken üzerinde sahip oldukları sonucu belirlemek için varyans testinin analizini kullanırlar.

### Tek Yönlü ve İki Yönlü Anova:

İki tür varyans analizi vardır: tek yönlü ve iki yönlü. Tek yönlü veya iki yönlü Varyans Analizi bağımsız değişkenlerin sayısını ifade eder. Tek yönlü ANOVA, tek bir bağımlı değişkeni etkileyen bir bağımsız değişkenin üzerindeki etkisini değerlendirir. Tüm örneklerin aynı olup olmadığını belirler. İki yönlü bir ANOVA, tek yönlü ANOVA'nın bir uzantısıdır. İki yönlü ANOVA'da, iki bağımsız değişken vardır. Örneğin, iki yönlü bir ANOVA, bir şirketin iki bağımsız değişkene dayalı işçi verimliliğini, maaş ve beceri seti olarak karşılaştırmasını sağlar. İki faktör arasındaki etkileşimi gözlemlemek için kullanılır. İki faktörün etkisini aynı anda test eder. (https://www.investopedia.com/terms/a/anova.asp)
Veri kümesinin bileşenleri üzerinde yapılan testlerde varyasyon vardır ve ANOVA bu varyasyonun sınıflandırma faktörü tarafından getirilen gruplandırma ile açıklanıp açıklanamayacağını araştırır. (https://explorable.com/anova)

#### ANOVA'nın R'de Uygulanması:

- anova(model)
- aov1 <- aov(class ~., data = dataset)   ~ Tüm değişkenlerle olan ilişkisini gösterir 
- summary(aov1)
- aov2 <- aov(class ~ age + bgr, data = dataset)    ~ Seçilen değişkenleri inceler 
- summary(aov2)


## Skewed Data:

Skewed data normal dağılıma sahip olmayan dengesiz, çarpık dağılımlı data anlamına gelir. Çarpıklık, verilerin simetri eksikliğini tanımlar. Bir dağılımın asimetri derecesinin ölçüsüdür. Çok değişkenli normallik, regresyonun tüm değişkenlerinin normal olmasını gerektirdiği anlamına gelir. Skew data ile normallik varsayımını ihlal ediyoruz.




### Çarpıklık Regresyon Modelini Nasıl Etkiler?

- Parametre tahminlerinde orantısız etkiye sebep olur. Parametre tahmini, kareli hatanın minimizasyonuna dayanır. Skewed datadaki gözlemler, parametre tahminleri üzerinde orantısız bir etki yapacaktır.

- Verilerin çarpıklığına bağlı olarak, Güven Aralıkları(istatistikler) normal dağılımlı hata varsayımına dayandığı için çok geniş veya çok dar olabilir.

- Çarpıklığı hesaba katacak şekilde özel olarak tasarlanmış modeller genellikle daha iyi performans gösterecektir. Bunlar, gama veya ters Gauss dağılımlarına sahip genelleştirilmiş doğrusal modelleri içerir.

### Logistic Regression Modelinde Skewed Bağımsız Değişkenin Sonuçları Nedir?

Lojistik regresyon bağımsız değişkenler için özel bir dağılım gerektirmez. Bunlar normal, çarpık, kategorik olabilir.
Bağımsız değişkenin normal dışı olması lineer gerileme uygulamaktan alıkoymaz, ancak bazı durumlarda sonuçların anlamlı olduğundan emin olmak için teşhislerde daha yakından dikkat etmek zorunda kalınabilir.

(https://www.quora.com/How-does-skewness-impact-regression-model)


## 2. Support Vector Machine (SVM):

Support Vector Machine bir sınıflandırma algoritmasıdır. SVM'in amacı, lineer programlama gibi gelişmiş optimizasyon metodlarını kullanarak verinin marjinini maksimize eden optimal ayırma hiperdüzlemini bulmaktır.
 (https://www.svm-tutorial.com/2014/11/svm-understanding-math-part-1/)

Etiketli veriler varsa SVM, veri alanını bölümlere ayırarak ve her segment sadece bir tür veri içerecek şekilde çoklu ayırma hiperdüzlemlerini üretmek için kullanılabilir. SVM tekniği, genellikle dağılımı bilinmeyen veriler anlamına gelen düzenli olmayan veriler için kullanışlıdır.                               
 (https://www.r-bloggers.com/machine-learning-using-support-vector-machines/)



Destek vektörleri, hiperdüzleme en yakın veri noktalarıdır, çıkarılmışsa, bölünen hiperdüzlemin konumunu değiştirecek bir veri kümesinin noktalarıdır. Bu nedenle, bir veri kümesinin kritik öğeleri olarak kabul edilebilirler. Hiperdüzlem basitçe, yukarıdaki grafikte olduğu gibi sadece iki özelliğe sahip bir sınıflandırma için, veri kümesini doğrusal olarak ayıran ve sınıflandıran bir çizgi olarak düşünülebilir. Hiperdüzlemden elde ettiğimiz veri noktalarının daha ötesinde, doğru bir şekilde sınıflandırılmış olduklarından daha emin olduk. Bu nedenle veri noktalarımızın mümkün olduğunca hiperdüzlemden uzakta olmasını ve yine de bunun doğru tarafında olmasını istiyoruz.

## 3. Karar Ağacı (Decision Tree):

Sınıflandırma için açık kurallar sağlayan ve heterojen veriler, eksik veriler ve doğrusal olmayan etkiler ile iyi başa çıkabilen veri madenciliğinin en sezgisel ve popüler yöntemlerinden biri karar ağacıdır. Öğeyle ilgili gözlemleri haritalandırarak bir öğenin hedef değerini tahmin eder. Doğrudan pazarlama, müşteri tutma, dolandırıcılık tespiti, tıbbi sorunların tanısı gibi konularda oldukça fazla kullanılır. (https://data-flair.training/blogs/r-decision-trees/)



Karar ağaçları sıklıkla insan düzeyini düşünmeyi taklit eder, böylece verileri anlamak ve bazı iyi yorumlar yapmak çok basittir.
Karar ağaçları, verilerin yorumlanması için mantığı gerçekten görmenizi sağlar (SVM, NN vb. gibi kara kutu algoritmaları gibi değil). Bir karar ağacı, her düğümün bir özelliği temsil ettiği bir ağaçtır. Her bir bağ (dal) bir kararı (kuralı) temsil eder ve her yaprak bir sonucu temsil eder (kategorik veya devam eder).

Bütün fikir, tüm veriler için bu şekilde bir ağaç oluşturmak ve her yaprakta tek bir sonucu işlemek (veya her yapraktaki hatayı en aza indirgemek).

(https://medium.com/deep-math-machine-learning-ai/chapter-4-decision-trees-algorithms-b93975f7a1f1)

Karar ağaçları, belirli bir hedef sınıfa olabildiğince saf olan alt kümeleri elde etmek için, eğitim setinin tekrarlı olarak bölünmesi yoluyla çalışır. Ağacın her düğümü, bir özellik üzerinde belirli bir testle bölünen belirli bir T kayıtları kümesine bağlanır. Örneğin, sürekli bir A niteliğinde bölünme test A≤x tarafından indüklenebilir. T kümeleri daha sonra ağacın sol dalına ve sağa doğru giden iki alt kümede bölünür.
#### Tl = { t ∈ T : t (A) ≤ x }     ve     Tr = { t ∈ T : t (A) > x }
Benzer şekilde, kategorik bir özellik B, ayrılmalarını kendi değerlerine göre uyarmak için kullanılabilir. Örneğin, 
#### B = {b1,…, bk}

Karar ağacını oluşturmak için özyinelemeli algoritmanın bölünme adımı, her bir özellik için olası tüm bölünmeleri dikkate alır ve seçilen bir kalite ölçüsüne göre en iyi olanı bulmaya çalışır: ayırma kriteri.
#### A1,…,An,C

Sürekli değişkenler için, karelerin toplamını en aza indirmek için her bir bölgeden bir sabit seçilir:

$$\min_{c\in \mathbb{R}} \sum_{x_i ∈ R_i} {(y_i-c)}^2 $$

Ni: Ri'deki gözlem sayısını belirtir. Sonuç olarak;

draw latex

$$\widehat{c_i} = \frac{1}{N_i} \sum_{x_k \in \mathbb{R_i}} y_k  $$

Benzer şekilde, çıktı kategorik olduğunda i düğümdündeki sınıf k gözlemlerinin oranı:


$$\widehat{p_{ik}} = \frac{1}{N_i}\sum_{x_l ∈ R_i}1_{y_l ∈ R_i} $$

Daha sonra, büyük çoğunluğa sahip gözlemler sınıflandırılırsa:


$$k(i) := \text{argmax} \widehat{p_{ik}}$$

### Gini Index:

Her nitelik ikiye bölünerek olası bütün ikiye bölünmelerin sınanmasıdır. Katsayı, 0 (%0) ile 1 (%100) aralığındadır; 0, mükemmel eşitliği temsil eder ve 1, mükemmel eşitsizliği temsil eder. Gini indeksi, rastgele seçilen bir ögenin ne sıklıkta yanlış tespit edildiğini ölçmek için kullanılan bir metriktir. Düşük gini indeksi olan bir özellik tercih edilmelidir. Gini indeksi kategorik hedef değişkeni için başarılı veya başarısız olarak çalışır. Gini indeks, yalnızca ikili bölmeleri ( 1 veya 0) gerçekleştirir ve yüksek gini indeksi homojenliği artırır. CART (Sınıflama ve Regresyon Ağacı) ikili bölmeler oluşturmak için gini yöntemini kullanır.

(https://veribilimcisi.com/2018/02/23/karar-agaclari-decision-trees/)

$$gini(S)=1-\sum_{j=1}^{n} p_j^2$$

gini indeks değerini veren ayrıma sahip değişken üzerinden bölünme gerçekleştirilir. Eğer

## 1. Logistic Regression:

```r
library(caTools)
library(e1071)
library(rpart)
```

Numerik verinin degisik kolonlari arasinda buyukluk farki var. Bazi kolonlar cok buyuk, bazilari cok ufak. Normalize etmek gerekli:

$$ z_i = \frac{x_i - \overline{x}}{\sqrt{var(x)}} $$

```r
kidney <- na.omit(read.csv("C:\\Users\\Lenova\\Desktop\\ch_kidney_disease.csv", na.strings = "?"))
for (name in setdiff(colnames(kidney),c("class"))) {
    kidney[,name] <- as.numeric(kidney[,name])
    kidney[,name] <- (kidney[,name] - mean(kidney[,name]))/sqrt(var(kidney[,name]))
}

kidney[,"class"] <- as.numeric(kidney[,"class"]) - 1
```

```r
set.seed(144)
```

```r
split <- sample.split(kidney$class, SplitRatio=3/10)  
training <-subset(kidney, split == T)
testing  <-subset(kidney, split == F)
```

```r
model <- glm(class~ age + bp + bgr + bu + sc + sod + pot + hemo + pcv + wbcc + rbcc, data = na.omit(training), family = binomial("logit"))
```

```r
summary(model)
```

```r
prediction <- predict(model, newdata = testing, type = "response")
confMat <- table(testing$class, prediction > 0.5)
```
